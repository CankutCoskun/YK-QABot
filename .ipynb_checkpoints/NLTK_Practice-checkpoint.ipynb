## 1.4   Reuters Corpus
The Reuters Corpus contains 10,788 news documents totaling 1.3 million words. The documents have been classified into 90 topics, and grouped into two sets, called "training" and "test"; thus, the text with fileid 'test/14826' is a document drawn from the test set

```python
import nltk
import nltk.data
from nltk.corpus import reuters
len(reuters.fileids())
```

```python
#Lancaster Stemmer
lancaster = nltk.LancasterStemmer()
tokens_Lanc_0 = [lancaster.stem(t) for t in doc0]
tokens_Lanc_0
```

### Lemmatization
The WordNet lemmatizer only removes affixes if the resulting word is in its dictionary.

```python
wnl = nltk.WordNetLemmatizer()
tokens_WN_0 = [wnl.lemmatize(t) for t in doc0]
print(len(tokens_WN_0))
tokens_WN_0
```
