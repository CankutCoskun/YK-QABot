```python
##Given a raw text 

def getText(raw_content):
    
    document = []
    
    for par in raw_content:
        if par.text != "" and not par.find("a"):
            document.append(par.text)
    
    return document

doc = getText(raw_content)

```

```python
for each in doc:
    print("###" + each)
```
