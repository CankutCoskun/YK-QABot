```python
import nltk 
from nltk.corpus import reuters
porter = nltk.PorterStemmer()
```

### Text Preprocessing
Porter Stemmer

```python
documents = []
for i in reuters.fileids():
    doc = [porter.stem(t) for t in reuters.words(i)]
    doc.insert(0,i)
    documents.append(doc)
```

```python
def createDict(documents):
    myDict = {}
    tf_table = []
    for doc in documents:
       
        docID = doc[0]
    
        for idx in range(len(doc)):
            if idx == 0:
                continue
                
            elif doc[idx] not in myDict.keys():
                myDict[doc[idx]] =[docID]
                
            else:
                myDict[doc[idx]].append(docID)
                
    return myDict
```

```python
newDict = createDict(documents)
newDict['export'][0:10]
```

### Search Engine

```python
class SearchEngine:
    
    myDict = {}
    
    def __init__(self,context):
        self.myDict = createDict(context)
        
    def search(self,inp):
        ste = porter.stem(inp)
        
        if ste not in self.myDict:
            print(inp, "does not exist in any document") 
        else:   
            print(self.myDict[ste][0:10])
    
```

```python
#prints the first ten document that the term appears in
myEngine = SearchEngine(documents)

myEngine.search('important')
```
